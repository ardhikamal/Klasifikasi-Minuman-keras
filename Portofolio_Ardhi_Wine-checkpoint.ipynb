```python
import pandas as pd

#Memproses dataset latihan
data = pd.read_csv("C:\Work\Data Science\My portofolio\Wine\Data minuman wine.csv")
data.head(10)
```

```python
# Membedakan features dan labels
features = ['Alcohol','Malic_acid','Ash','Alcalinity','Magnesium','Phenols','Flavanoids','Nonflavanoids','Proanthocyanins','Color_intensity','Hue','OD280_315_of_diluted_wines','Proline']
label = 'WineVariety'
X, y = data[features].values, data[label].values

for n in range(0,5):
    print("Minuman Wine Ke-", str(n+1), "\n  Features:",list(X[n]), "\n  Label:", y[n])
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

for col in features:
    data.boxplot(column=col, by=label, figsize=(7,7))
    plt.title(col)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split data 70%-30% ke set latihan dan set tes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print (' Kasus Training : %d\n Kasus Test : %d' % (X_train.shape[0], X_test.shape[0]))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Mendefinisikan preprocessing untuk kolom numerik (skala mereka)
feature_columns = [0,1,2,3,4,5,6]
feature_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
    ])

# Membuat step sebelum processing
preprocessor = ColumnTransformer(
    transformers=[
        ('preprocess', feature_transformer, feature_columns)])

# Membuat pipeline latihan
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('regressor', LogisticRegression(solver='lbfgs', multi_class='auto'))])


# Menyesuaikan pipeline untuk melatih model regresi linear pada set latihan
model = pipeline.fit(X_train, y_train)
print (model)
```

```python
from sklearn. metrics import accuracy_score, precision_score, recall_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Mendapatkan prediksi dari tes data
predictions = model.predict(X_test)

# Mendapatkan metrik
print("Akurasi Keseluruhan:",accuracy_score(y_test, predictions))
print("Presisi Keseluruhan:",precision_score(y_test, predictions, average='macro'))
print("Recall Keseluruhan:",recall_score(y_test, predictions, average='macro'))

# Menggambar confusion matrix
cm = confusion_matrix(y_test, predictions)
classes = ['Varietas A','Varietas B','Varietas C']
plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
plt.colorbar()
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)
plt.title('Confusion Matrix')
plt.xlabel("Varietas Prediksi")
plt.ylabel("Varietas yang sebenarnya")
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Mendapatkan Score Kemungkinan kategori
probabilities = model.predict_proba(X_test)

auc = roc_auc_score(y_test,probabilities, multi_class='ovr')
print('Rerata AUC:', auc)

# Mendapatkan metrik ROC untuk masing-masing kategori
fpr = {}
tpr = {}
thresh ={}
for i in range(len(classes)):    
    fpr[i], tpr[i], thresh[i] = roc_curve(y_test, probabilities[:,i], pos_label=i)
    
# Menggambar grafik ROC
plt.plot(fpr[0], tpr[0], linestyle='--',color='orange', label=classes[0] + ' vs Sisanya')
plt.plot(fpr[1], tpr[1], linestyle='--',color='green', label=classes[1] + ' vs Sisanya')
plt.plot(fpr[2], tpr[2], linestyle='--',color='blue', label=classes[2] + ' vs Sisanya')
plt.title('Kurva ROC Multi Kategori')
plt.xlabel('Tingkat Positif palsu')
plt.ylabel('Tingkat Positif benar')
plt.legend(loc='best')
plt.show()
```
